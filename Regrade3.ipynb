Regra de três simples

```python
grandeza1=input("Insira o nome da grandeza que tem todos os valores conhecidos:")
grandeza2=input("Insira o nome da grandeza que apresenta um valor desconhecido:")
```

```python
a=float(input(f"Insira o valor da grandeza '{grandeza1}' que possui valor conhecido correspondente na grandeza '{grandeza2}':"))
b=float(input(f"Insira o valor correspondente da {grandeza2}':"))
```

```python
c=float(input(f"Insira o valor da grandeza '{grandeza1}' que não possui valor conhecido correspondente na grandeza '{grandeza2}':"))
```

```python
print(f"Essas grandezas, {grandeza1} e {grandeza2} são diretamente (D) ou inversamente proporcionais (I)?")
G = input('Digite D ou I:').upper()

if G == "D":
    print("Você escolheu Diretamente Proporcionais")
elif G == "I":
    print("Você escolheu Inversamente Proporcional")
else:
    print("Entrada inválida. Digite D ou I.")
```

if, elif e else formam uma estrutura de decisão encadeada.

if define a primeira condição a ser testada.

elif (“senão, se”) define condições alternativas, testadas em sequência, somente se as anteriores falharem.

else define o caso restante, isto é, o que acontece quando nenhuma das condições anteriores foi atendida.

else não testa condição; ele é o “resto”.

```python
if G == "D":
  x = (c*b)/a
  print(f"O valor da grandeza '{grandeza2}' é {x}")

else:
  x = (a*b)/c
  print(f"O valor da grandeza '{grandeza2}' é {x}")
```
