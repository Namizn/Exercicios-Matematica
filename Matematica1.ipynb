```python
a = int (input("Escolha um numero positivo inteiro: "))
b = int(input("Escolha outro numero positivo inteiro: "))

a = int(a)
b = int(b)

valor = a + b

print("A soma dos valores é", valor)
```

```python
#Calculadora
num1 = float(input("Escolha um numero "))
num2 = float(input("Escolha outro numero "))

print ("soma: ", num1 + num2)
print ("subtração: ", num1 - num2)
print ("multiplicação: ", num1 * num2)
print ("divisão: ", num1 / num2)
print ("divisão com duas casas decimais: ", round(num1/num2, 2))
print ("parte de divisão inteira: ", num1 // num2)
print ("resto da divisão: ", num1 % num2)
print ("potência: ", num1 ** num2)
print ("raíz quadrada: ", num1 ** 0.5)
print ("raíz cubica: ", num1 ** (1/3))
```
